#### .children

I belive that this generally serves the same purpose, but the **`.children`** approach isn't a `list`. This still reads the newline `'\n'` but doesn't display it. Also still loads the comment in a weird way.
They way the comments are loaded may just be a consequence of the 'lxml-xml' parser

According to Bard-AI:
> The main difference between `.children` and `.contents` is that `.children` returns a **generator**, while while `.contents` returns a **list** 

> A **generator** is an iterable object that *produces its elements one at a time*, rather than storing them all in memory at once. This makes it a more efficient way to handle large amounts of data, as it does not require all of the dat to be loaded into memory at once.

The above is likely the case why I lose the information after running the `.children` output through a function such as the len()

```python
soupBodyChildren = soup1.body.children

# print(len(list(soupBodyChildren)))

for code_line in soupBodyChildren:
    if code_line == '\n':
        pass
    else: 
        print("+ " + str(code_line))
```

#### .descendants

From the documentation:
> The `.descendants` attribute lets you iterate over *all* of a tag’s children, recursively: its direct children, the children of its direct children, and so on:

The `.contents` and the `.children` approach only views the *direct* children as an "element" in the *list* or *generator.* The `.descendants` approach includes the *children of the children* as "elements".

**NOTE**: The *grandchildren* are their own *element,* but they are still present WITHIN the *element* of their parent (SEE BELOW)

```python
soupBodyDescendants = soup1.body.descendants

# print(len(list(soupBodyDescendants)))

for code_line in soupBodyDescendants:
    if code_line == '\n':
        pass
    else: 
        print("+ " + str(code_line))
```

#### A function

In case the matching system provided by the re library and other suggested approaches of the Beautiful Soup documentation aren't enough for a specific use case, you can simply create your own function.

Conditions set by the bSoup documentation:
- The element should be the only argument accepted by the function
- If the condition matches, return `True`, return `False` otherwise

```python
# The function returns `True` if the folowing conditions are met
# The benefit of using a function is the ability to FILTER OUT conditions and adding complexity.

# What's speacial about the functions is that we're using the .has_attr method

def has_class_but_no_id(tag_var):
    return tag_var.has_attr('class') and not tag_var.has_attr('id')

def has_id_but_no_class(tag_var):
    return tag_var.has_attr('id') and not tag_var.has_attr('class')

def has_class_and_id(tag_var):
    return tag_var.has_attr('id') and tag_var.has_attr('class')

soup1.find_all(has_class_but_no_id) # replace the argument with the above functions to see the differences

# Apparently, you can pass functions as arguments to another function
# Honestly, I'm not sure exactly how the logic follows here
# Perhaps, the .find_all() function is set-up specifically work in this way
# I think I should just follow the format of the function given above
```

#### .extract()

```python
# removes a tag or string FROM the TREE
# and stores it into a new variable for use later
# It's essentially like a CUT for HTML code

markup = '<a href="http://example.com/">I linked to <i>example.com</i></a>'
soupy = bSoup(markup, 'html.parser')
a_tag = soupy.a

i_tag = soupy.i.extract()

print(a_tag)
print()
# <a href="http://example.com/">I linked to</a>

print(i_tag)
print()
# <i>example.com</i>

print(i_tag.parent)
print()
# None

my_string = i_tag.string.extract()
print(my_string)
print()
# 'example.com'

print(my_string.parent)
print()
# None
print(i_tag)
print()
# <i></i>
```

#### .smooth()

```python
# Notice that, whenever we appended, it created a new ELEMENT into something that \
#   resembles a list 
# The .smooth() combines these invididual textual elements into a single string.

soupy = bSoup("<p>A one</p>", 'html.parser')
soupy.p.append(", a two")

soupy.p.contents
# ['A one', ', a two']

print(soupy.p.encode())
print()
# b'<p>A one, a two</p>'

print(soupy.p.prettify())
print()
# <p>
#  A one
#  , a two
# </p>

soupy.smooth()

print(soupy.p.contents)
print()
# ['A one, a two']

print(soupy.p.prettify())
# <p>
#  A one, a two
# </p>
```
